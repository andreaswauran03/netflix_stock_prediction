```python
import pandas as pd
import math 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense, LSTM
from keras.models import Sequential
```

```python
df = pd.read_csv("NFLX.csv")
df.head()
```

```python
# Info
df.info()
```

```python
# Checking Duplicated
df.duplicated().sum()
```

```python
# Convert Date column type
df["Date"] = pd.to_datetime(df["Date"])
```

```python
# Filtered Date
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2021-12-31')

df = df[(df["Date"] >= start_date) &
        (df["Date"] <= end_date)]
df = df.reset_index(drop=True)
df.head()
```

```python
df.describe()
```

```python
# Visualized Data
plt.figure(figsize=(10,6))
plt.plot(df["Date"],
        df["Close"])
plt.xlabel("Date", fontsize=10)
plt.ylabel("Price ($)", fontsize=10)
plt.title("Closing Price in Netflix")
plt.show()
```

```python
# Filtering Dataset
data = df.filter(["Close"]).values
# Create number of training dataset
training_data_len = math.ceil(len(data) * 0.8)
training_data_len
```

```python
# Scaled Dataset
mms = MinMaxScaler()
scaled_data = mms.fit_transform(data)
scaled_data
```

```python
# Create Train Dataset
train_dataset = scaled_data[:training_data_len]
```

```python
# Making prediction for 50 days
x_train = []
y_train = []

for i in range(30, len(train_dataset)):
    x_train.append(train_dataset[i-30:i, 0])
    y_train.append(train_dataset[i, 0])
```

```python
# List to Array
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshape Array
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# Create Model
lstm = Sequential()
lstm.add(LSTM(30, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
lstm.add(LSTM(30, return_sequences=False))
lstm.add(Dense(10))
lstm.add(Dense(1))
```

```python
# Compile Model
lstm.compile(optimizer="adam",
            loss="mean_squared_error")
```

```python
lstm.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
# Create Test Dataset
# Dataset from 556-606
test_data = scaled_data[training_data_len-30:, :]
# Create x_test and y_test
x_test = []
y_test = data[training_data_len:, :]

for i in range(30, len(test_data)):
    x_test.append(test_data[i-30:i, 0])
```

```python
# List to Array
x_test = np.array(x_test)
```

```python
x_test.shape
```

```python
# Reshape the data
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
# Making Predictions
y_pred = lstm.predict(x_test)
y_pred = mms.inverse_transform(y_pred)
```

```python
# RMSE (Root Mean Squared Error)
rmse = np.sqrt( np.mean(y_pred - y_test)**2 )
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# Plot the Data
train = df[:training_data_len]
valid = df[training_data_len:]
valid["Prediction"] = y_pred
```

```python
# Visualized the data
plt.figure(figsize=(10, 6))
plt.title("Model")
plt.plot(train["Close"])
plt.plot(valid[["Close", "Prediction"]])
plt.xlabel("Date", fontsize=18)
plt.ylabel("Close Price USD ($)", fontsize=18)
plt.legend(["Train", "Val", "Prediction"])
plt.show()
```

```python
# Prediction latest price
# Import dataset
df = pd.read_csv("NFLX.csv")
data = df.filter(["Close"])
# Create X_dataset
last_30d = data[-30:].values
#Transform dataset
last_30d_scaled = mms.transform(last_30d)
x_test = []
x_test.append(last_30d_scaled)
# List to Array
x_test = np.array(x_test)
# Reshape Array
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
# Making prediction
y_pred = lstm.predict(x_test)
# Unscaled dataset
y_pred = mms.inverse_transform(y_pred)
```

```python
predict = pd.DataFrame(
    {"Actual": df["Close"].iloc[-1],
    "Prediction": y_pred[0][0]},
    index=[0]
)

predict
```
